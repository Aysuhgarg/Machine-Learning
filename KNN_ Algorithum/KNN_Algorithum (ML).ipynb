```python
mean_01=np.array([3.0,4.0])
#print(mean_01)
cov_01=np.array([[1.0,0.5],[0.5,1.0]])
dist_01=np.random.multivariate_normal(mean_01,cov_01,200)
print (dist_01.shape)
print(dist_01)

mean_02=np.array([0.0,-1.0])
#print(mean_01)
cov_02=np.array([[0.45,0.5],[0.5,0.6]])
dist_02=np.random.multivariate_normal(mean_02,cov_02,200)
print (dist_02.shape)
print(dist_02)
```

```python
plt.figure(0)
for x in range(dist_01.shape[0]):
    plt.scatter(dist_01[x,0],dist_01[x,1],color='red')
    plt.scatter(dist_02[x,0],dist_02[x,1],color='green')
plt.show()
```

```python
labels=np.zeros((400,1))
labels[200:]=1.0
x_data=np.zeros((400,2))
x_data[:200, : ]=dist_01
x_data[200:, : ]=dist_02
print(x_data)
print(labels)
```

```python
# KNN_Algorithum
def dist(x1,x2):
    return np.sqrt(((x1-x2)**2).sum())

x1=np.array([0.0,0.0])
x2=np.array([1.0,1.0])
print(dist(x1,x2))
```

```python
def knn(x_train,query_point,y_train,k=5):
    cals=[]
   # updated_cals=[]
    for i in range(x_train.shape[0]):
        v=[dist(query_point,x_train[i, : ]),y_train[i]]
        cals.append(v)
    updated_cals=sorted(cals)
    pred_arr=np.asarray(updated_cals[ :k])
    pred_arr=np.unique(pred_arr[ : ,1],return_counts=True)
    
    index=pred_arr[1].argmax()
    return pred_arr[0][index]
q=np.array([2.0,0.0])
predicted_label=knn(x_data,q,labels)
print(predicted_label)
```
